Step 1:Preprocess the tweets

```python
import re
```

```python
#nltk-natural language toolkit
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
stopword.extend(['US','RT','rt'])
nltk.download('words')
```

```python
stopword
```

```python
def preprocess(text):
  text = re.sub(r'http\S+', '', text)
  text = re.sub(r"\W", '',text) #keeping only text and digit
  text = re.sub(r"\d", '',text) #removing the digits
  text = re.sub(r"_", '',text)
  text = text.lower()
  text = re.split('\W+', text)
  text = [word for word in text if word not in stopword]
  text = "".join(text)
  return text
```

```python
tweet_data['cleaned'] = tweet_data['tweet'].apply(lambda x:preprocess(x))
```

Sentiment

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
  blob = TextBlob(text)
  return blob.sentiment.polarity #polarity-(positive or negative)
```

```python
tweet_data['sentiment'] = tweet_data['cleaned'].apply(lambda x: get_sentiment(x))
```
